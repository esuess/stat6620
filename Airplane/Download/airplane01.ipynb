# Using wget

First download 3 files to see how *wget* works.

```bash
wget http://stat-computing.org/dataexpo/2009/1987.csv.bz2
```

```bash
wget http://stat-computing.org/dataexpo/2009/1988.csv.bz2
```

```bash
wget http://stat-computing.org/dataexpo/2009/1989.csv.bz2
```

Now we will time downloading all of the files.

The -nv is for no verbose, so there is limited output from *wget*.

```bash
time wget -nv http://stat-computing.org/dataexpo/2009/{1987..2008}.csv.bz2
```

Notice the use of the *seq* function.

```bash
time seq 1987 2008 | parallel wget -nv http://stat-computing.org/dataexpo/2009/{}.csv.bz2
```

Now we use the ::: for input into the *parallel* function. 

```bash
time parallel wget -nv http://stat-computing.org/dataexpo/2009/{}.csv.bz2 ::: {1987..2008}
```

# unzip the data using bunzip

The -k option keeps the orginal files.

```bash
bunzip2 -k 19**.csv.bz2
bunzip2 -k 200*.csv.bz2
```

```bash
time bunzip2 -k {1987..2008}.csv.bz2
```

-v option  verbose

-f option  force overwrite

-k option  keep original files

```bash
time parallel bunzip2 -v -f -k {}.csv.bz2 ::: {1987..2008}
```

-dv option  decompress, verbose

-r option   reads the entire file into memory and split between processors

-p16 option gives 16 processors, the number should be the number of processor cores

-f option   force overwrite

-k option   keep original files

Question: is pbzip already parallel?  Do you need the parallel call?

```bash
time parallel pbzip2 -dv -r -p8 -f -k {}.csv.bz2 ::: {1987..2008}
```

# using *csvkit*

```bash
csvlook 1987.csv | head
csvlook 1987.csv.bz2
```

```bash
csvstack {1987..1999}.csv > output1900.csv
csvstack {2000..2008}.csv > output2000.csv

csvstack {1987..2008}.csv > output.csv
```

```bash
# check file encoding
# file -bi <filename>

file --mime-encoding 200*.csv
iconv -l
iconv -f ISO-8859-1 -c -t us-ascii 2001.csv > 2001a.csv
iconv -f ISO-8859-1 -c -t us-ascii 2002.csv  > 2002a.csv

cp 2001a.csv 2001.csv
cp 2002a.csv 2002.csv

file --mime-encoding 200*.csv

time csvstack {1987..1999}.csv > output1900.csv
time csvstack {2000..2008}.csv > output2000.csv

time csvstack {1987..2008}.csv > output.csv

time csvcut -c DayOfWeek output.csv | grep 6 | wc -l
```
